```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
iq = np.array([1.5,2.4,3.7,4.7 , 7.85,6.5,7.2,8.6,9.5,10.0])
placement =np.array([2, 4, 3.4, 7,6,8 , 9, 10, 11,12])
```

```python
plt.scatter(iq, placement)
```

```python
meanx=np.mean(iq)
print(meanx)
```

```python
meany = np.mean(placement)
print(meany)
```

Function for finding the m and b

```python
rows = iq.shape[0]
print(rows)
```

```python
m= 0
b = 0

num= 0
den = 0
for i in range(rows):
  num += (placement[i]-meany)  * (iq[i]- meanx)
  den += (iq[i]- meanx)**2
m = num/den
b = meany - (m*meanx)
print(m,b)
```
